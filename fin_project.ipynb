10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

```python
left = df[df['left'] == 1 ]['satisfaction_level']
remained = df[df['left'] == 0 ]['satisfaction_level']
stats.ttest_ind(left, remained)
```

```python
sns.distplot(left, kde_kws={"label": "left"})
sns.distplot(remained, kde_kws={"label": "remained"})
fig=plt.gcf()
fig.set_size_inches(11,7)
plt.show()
```

Очевидно, что уходят соотрудники, у которых уровень удовлетворения работой - ниже, но важно узнать о чего зависит satisfaction_level

```python
fig = plt.figure(figsize=(20,20))
cols = ['satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident', 'left',
       'promotion_last_5years', 'salary']
count = 1
for col in cols[1:]:
    plt.subplot(4, 2 , count)
    sns.barplot(x=col, y='satisfaction_level', data=df)
    count+=1
```

Анализирую график, можно сделать следующие выводы:
    1. Чем больше среднее кол-во часов в месяц тем меньше уровень удовлетворения ( переработки пагубно влияют)
    2. Так же вляет кол-во проектов ( вероятно кол-во проектов связано с кол-вом часов в месяц)
    3. Остальные показатели стоит рассмотреть детальней с разбивкой по департаментам

Сначала проверим взаимосвязь number_project и average_montly_hours

```python
sns.barplot(df.average_montly_hours, df.number_project)
plt.show()
```

Получили достаточно очевидный результат, чем больше проектов, тем больше переработок.
Рассмотрим другие показатели в разрезе департаментов

```python
fig = plt.figure(figsize=(20,20))
cols = ['Work_accident','promotion_last_5years', 'salary']
count = 1
for col in cols:
    plt.subplot(3, 1 , count)
    sns.barplot(x=col, y='satisfaction_level', data=df, hue = 'department')
    count+=1
```

Как видим, особых различий не наблюдается. Попробуем посмотреть напрямую с классом left

```python
fig = plt.figure(figsize=(20,20))
cols = ['Work_accident','promotion_last_5years', 'salary']
count = 1
for col in cols:
    plt.subplot(3, 1 , count)
    sns.barplot(x='left', y=col, data=df, hue = 'department')
    count+=1
```

Уже здесь, можно сделать более точечный вывод, что в отеделе managment продвижение по карьерной лестнице и зарплата достаточно сильно влияют на то йудет ли соотрудник с компании или нет

Теперь рассмотрим детальнее кол-во часов в месяц

```python
fig = plt.figure(figsize=(10,10))
sns.barplot(df.average_montly_hours, df.left)
plt.show()
```

Как видно из графика, ситуация не так однозначна. Слишком малое кол-во часов в месяц так же влияет на уход соотрудника из компании

```python
fig = plt.figure(figsize=(8,8))
plt.hist(df.average_montly_hours, bins = 30)
plt.show()
```

Видно что распределение имеет перекор левее и правее центра. Вероятно при нормализации ситуации с кол-вом часов ( приведению распределения к нормальному) кол-во соотрудников уходящих из компании снизится

Резюмирую:<br> Можно сказать что главным фактором ухода из компании является число переработок, так же влияет уровень зарплата в отдельных департаментах ( managment).<br>
Кол-во переработок среди персонала с низкой зарплатой ( персонал более низкой категории) может быть связано с неправильно выстроенной системой управления.<br> Т.е не выстроены бизнес-процессы и для осуществления простых операций нужно затрачивать больше сил. Это так же влияет и на управленческое звено, т.к без отлаженной системы выстраивать работу в команде проблематично.
